```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load unified dataset
unified_df = pd.read_csv('unified_dataset_with_task_labels.csv')

# Separate the execution times and communication costs from the unified dataset
execution_times = unified_df[['VM1 Execution Time', 'VM2 Execution Time', 'VM3 Execution Time']].to_numpy()

# Add placeholder execution times for VM4 and VM5 if needed
def extend_execution_times(execution_times, num_vms):
    if num_vms > execution_times.shape[1]:
        extended_times = np.hstack([execution_times, np.full((execution_times.shape[0], num_vms - execution_times.shape[1]), np.max(execution_times))])
        return extended_times
    return execution_times

comm_costs = unified_df[['Comm Cost to Task 1', 'Comm Cost to Task 2', 'Comm Cost to Task 3', 
                         'Comm Cost to Task 4', 'Comm Cost to Task 5']].to_numpy()

# Task Dependency Graph (DAG) - Modify based on your actual task dependencies
G = nx.DiGraph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 5)])

# First-Come-First-Serve (FCFS) Scheduling
def fcfs_scheduling(execution_times, num_vms):
    schedule = {}
    vm_finish_time = [0] * num_vms  # Tracks when each VM finishes
    for task in range(len(execution_times)):
        best_vm = min(range(num_vms), key=lambda vm: vm_finish_time[vm])
        schedule[task + 1] = best_vm
        vm_finish_time[best_vm] += execution_times[task][best_vm]
    return schedule, vm_finish_time

# Round Robin Scheduling
def round_robin_scheduling(execution_times, num_vms):
    schedule = {}
    vm_finish_time = [0] * num_vms
    task_queue = list(range(1, len(execution_times) + 1))  # Task IDs start from 1

    while task_queue:
        for vm in range(num_vms):
            if task_queue:
                task = task_queue.pop(0)  # Get the next task
                schedule[task] = vm
                vm_finish_time[vm] += execution_times[task - 1][vm]  # Update finish time for the assigned VM

    return schedule, vm_finish_time

# Shortest Job First Scheduling
def sjf_scheduling(execution_times, num_vms):
    schedule = {}
    vm_finish_time = [0] * num_vms
    task_indices = list(range(len(execution_times)))

    # Sort tasks based on minimum execution time on any VM
    sorted_tasks = sorted(task_indices, key=lambda t: np.min(execution_times[t]))

    for task in sorted_tasks:
        best_vm = min(range(num_vms), key=lambda vm: vm_finish_time[vm] + execution_times[task][vm])
        schedule[task + 1] = best_vm
        vm_finish_time[best_vm] += execution_times[task][best_vm]
    
    return schedule, vm_finish_time

# K-Means Clustering for Task Prioritization
def cluster_tasks(execution_times, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(execution_times)
    return kmeans.labels_  # Task labels after clustering

# HEFT (Heterogeneous Earliest Finish Time) Algorithm for VM assignment
def heft_scheduling(task_graph, execution_times, comm_costs, num_vms):
    def calc_upward_rank(task, exec_times, comm_costs):
        successors = list(task_graph.successors(task))
        if not successors:
            return np.mean(exec_times[task - 1])  # If no successors, upward rank is the execution time
        return np.mean(exec_times[task - 1]) + max(calc_upward_rank(t, exec_times, comm_costs) for t in successors)

    # Step 1: Prioritize tasks using upward rank
    task_priority = {task: calc_upward_rank(task, execution_times, comm_costs) for task in task_graph.nodes()}
    sorted_tasks = sorted(task_priority, key=task_priority.get, reverse=True)
    
    # Step 2: Assign tasks to VMs using the HEFT approach
    schedule = {}
    vm_finish_time = [0] * num_vms  # Finish time of each VM
    for task in sorted_tasks:
        best_vm = min(range(num_vms), key=lambda vm: vm_finish_time[vm] + execution_times[task - 1][vm])
        vm_finish_time[best_vm] += execution_times[task - 1][best_vm]
        schedule[task] = best_vm
    return schedule, vm_finish_time

# Evaluate Performance
def evaluate_schedule(schedule, execution_times):
    vm_finish_times = [0] * num_vms
    for task, vm in schedule.items():
        vm_finish_times[vm] += execution_times[task - 1][vm]
    makespan = max(vm_finish_times)
    speedup = sum(np.min(execution_times, axis=1)) / makespan
    efficiency = speedup / len(set(schedule.values()))  # Number of VMs used
    return makespan, speedup, efficiency

# PBKMS Algorithm: Combines K-means with Scheduling (Previously K-means HEFT Hybrid)
def pbkms_scheduling(execution_times, comm_costs, num_vms):
    clustered_tasks = cluster_tasks(execution_times, num_vms)  # Cluster tasks
    return heft_scheduling(G, execution_times, comm_costs, num_vms)

# Visualize Results
def plot_results(vm_counts, makespan_results, speedup_results, efficiency_results):
    plt.figure(figsize=(10, 6))
    
    # Makespan Plot
    plt.subplot(1, 3, 1)
    plt.plot(vm_counts, [result[0] for result in makespan_results], marker='o', label='FCFS', color='blue')
    plt.plot(vm_counts, [result[1] for result in makespan_results], marker='o', label='Round Robin', color='orange')
    plt.plot(vm_counts, [result[2] for result in makespan_results], marker='o', label='SJF', color='green')
    plt.plot(vm_counts, [result[3] for result in makespan_results], marker='o', label='PBKMS', color='red')
    plt.title('Makespan vs VMs')
    plt.xlabel('Number of VMs')
    plt.ylabel('Makespan')
    plt.legend()

    # Speedup Plot
    plt.subplot(1, 3, 2)
    plt.plot(vm_counts, [result[0] for result in speedup_results], marker='o', label='FCFS', color='blue')
    plt.plot(vm_counts, [result[1] for result in speedup_results], marker='o', label='Round Robin', color='orange')
    plt.plot(vm_counts, [result[2] for result in speedup_results], marker='o', label='SJF', color='green')
    plt.plot(vm_counts, [result[3] for result in speedup_results], marker='o', label='PBKMS', color='red')
    plt.title('Speedup vs VMs')
    plt.xlabel('Number of VMs')
    plt.ylabel('Speedup')
    plt.legend()

    # Efficiency Plot
    plt.subplot(1, 3, 3)
    plt.plot(vm_counts, [result[0] for result in efficiency_results], marker='o', label='FCFS', color='blue')
    plt.plot(vm_counts, [result[1] for result in efficiency_results], marker='o', label='Round Robin', color='orange')
    plt.plot(vm_counts, [result[2] for result in efficiency_results], marker='o', label='SJF', color='green')
    plt.plot(vm_counts, [result[3] for result in efficiency_results], marker='o', label='PBKMS', color='red')
    plt.title('Efficiency vs VMs')
    plt.xlabel('Number of VMs')
    plt.ylabel('Efficiency')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Testing with various VM counts
vm_counts = [2, 3, 4, 5]
makespan_results = []
speedup_results = []
efficiency_results = []

for num_vms in vm_counts:
    print(f"\nTesting with {num_vms} VMs")
    
    # Extend execution times if required
    extended_execution_times = extend_execution_times(execution_times, num_vms)
    
    # Run FCFS Algorithm
    fcfs_schedule, _ = fcfs_scheduling(extended_execution_times, num_vms)
    fcfs_makespan, fcfs_speedup, fcfs_efficiency = evaluate_schedule(fcfs_schedule, extended_execution_times)
    
    # Run Round Robin Algorithm
    rr_schedule, _ = round_robin_scheduling(extended_execution_times, num_vms)
    rr_makespan, rr_speedup, rr_efficiency = evaluate_schedule(rr_schedule, extended_execution_times)
    
    # Run Shortest Job First Algorithm
    sjf_schedule, _ = sjf_scheduling(extended_execution_times, num_vms)
    sjf_makespan, sjf_speedup, sjf_efficiency = evaluate_schedule(sjf_schedule, extended_execution_times)
    
    # Run PBKMS Algorithm
    pbkms_schedule, _ = pbkms_scheduling(extended_execution_times, comm_costs, num_vms)
    pbkms_makespan, pbkms_speedup, pbkms_efficiency = evaluate_schedule(pbkms_schedule, extended_execution_times)
    
    # Print Results
    print(f"FCFS Makespan: {fcfs_makespan}, Speedup: {fcfs_speedup}, Efficiency: {fcfs_efficiency}")
    print("FCFS Task Assignment:", fcfs_schedule)
    
    print(f"Round Robin Makespan: {rr_makespan}, Speedup: {rr_speedup}, Efficiency: {rr_efficiency}")
    print("Round Robin Task Assignment:", rr_schedule)
    
    print(f"Shortest Job First Makespan: {sjf_makespan}, Speedup: {sjf_speedup}, Efficiency: {sjf_efficiency}")
    print("Shortest Job First Task Assignment:", sjf_schedule)
    
    print(f"PBKMS Makespan: {pbkms_makespan}, Speedup: {pbkms_speedup}, Efficiency: {pbkms_efficiency}")
    print("PBKMS Task Assignment:", pbkms_schedule)

    # Collect results for visualization
    makespan_results.append([fcfs_makespan, rr_makespan, sjf_makespan, pbkms_makespan])
    speedup_results.append([fcfs_speedup, rr_speedup, sjf_speedup, pbkms_speedup])
    efficiency_results.append([fcfs_efficiency, rr_efficiency, sjf_efficiency, pbkms_efficiency])

# Plot Results
plot_results(vm_counts, makespan_results, speedup_results, efficiency_results)
```
